```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("./datasets/traffic_collisions.csv")
```

```python
df.columns
```

```python
mapSizeX = 12
mapSizeY = 10
```

```python
plt.figure(figsize=(mapSizeX, mapSizeY))
plt.scatter(df['LONG'], df['LAT'], s=1, alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Toronto Collision Locations')
plt.show()
```

```python
injury_fatal_df = df[(df['FATALITIES'] > 0) | (df['INJURY'] == "YES")]
plt.figure(figsize=(mapSizeX, mapSizeY))
plt.scatter(injury_fatal_df['LONG'], injury_fatal_df['LAT'], s=1, alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Toronto FATAL Collision Locations')
plt.show()
```

```python
injury_fatal_df.shape
```

```python
pedestrian_df = df[(df['PEDESTRIAN'] == "YES")]
plt.figure(figsize=(mapSizeX, mapSizeY))
plt.scatter(pedestrian_df['LONG'], pedestrian_df['LAT'], s=1, alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Toronto PEDESTRIAN Collision Locations')
plt.show()
```

```python
pedestrian_df.shape
```

```python
bicycle_df = df[(df['BICYCLE'] == "YES")]
plt.figure(figsize=(mapSizeX, mapSizeY))
plt.scatter(bicycle_df['LONG'], bicycle_df['LAT'], s=1, alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Toronto Bicycle Collision Locations')
plt.show()
```

```python
bicycle_df.shape
```

```python
automobile_pedestrian_df = df[(df['PEDESTRIAN'] == "YES") & (df['AUTOMOBILE'] == "YES") ]
plt.figure(figsize=(mapSizeX, mapSizeY))
plt.scatter(automobile_pedestrian_df['LONG'], automobile_pedestrian_df['LAT'], s=1, alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Toronto Automobile-Pedestrian Collision Locations')
plt.show()
```

```python
automobile_pedestrian_df.shape
```

```python
motorcycle_df = df[(df['MOTORCYCLE'] == "YES")]
plt.figure(figsize=(mapSizeX, mapSizeY))
plt.scatter(motorcycle_df['LONG'], motorcycle_df['LAT'], s=1, alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Toronto MOTORCYCLE Collision Locations')
plt.show()
```

```python
motorcycle_df.shape
```

```python
automobile_motorcycle_df = df[(df['MOTORCYCLE'] == "YES") & (df['AUTOMOBILE'] == "YES")]
plt.figure(figsize=(mapSizeX, mapSizeY))
plt.scatter(automobile_motorcycle_df['LONG'], automobile_motorcycle_df['LAT'], s=1, alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Toronto MOTORCYCLE Collision Locations')
plt.show()
```

```python
automobile_motorcycle_df.shape
```

```python
order = sorted(pedestrian_df['DIVISION'].unique())
sns.countplot(x="DIVISION", data=df, order=order)
```
