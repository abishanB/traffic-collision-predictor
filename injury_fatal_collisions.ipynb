```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("./datasets/traffic_collisions.csv")
```

```python
df.head(2)
```

```python
mapSizeX = 12
mapSizeY = 10
```

```python
plt.figure(figsize=(mapSizeX, mapSizeY))
plt.scatter(df['LONG'], df['LAT'], s=1, alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Toronto Collision Locations')
plt.show()
```

```python
injury_df = df[(df['INJURY'] == "YES")]
plt.figure(figsize=(mapSizeX, mapSizeY))
plt.scatter(injury_df['LONG'], injury_df['LAT'], s=1, alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Toronto FATAL Collision Locations')
plt.show()
```
